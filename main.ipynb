#### 1. Import pandas library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2. Import users table:

```python
users_df = pd.read_csv("users_table.csv")
users_df.head()
```

#### 3. Rename Id column to userId

```python
users_df = users_df.rename(columns={"Id": "user_Id"})
```

#### 4. Import posts table:

```python
posts_df = pd.read_csv("posts_table.csv")
posts_df.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts_df = posts_df.rename(columns={"Id": "post_Id", "OwnerUserId": "user_Id"}) 
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
Users_columns_df = users_df[['user_Id', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
posts_columns_df = posts_df[['post_Id', 'user_Id', 'Score', 'ViewCount', 'CommentCount']]
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
new_df = pd.merge(Users_columns_df, posts_columns_df, on= "user_Id", how="inner")
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
"""
with inner join, we only get the rows where there is a match in both dataframes.
so if there are any user_Id in the posts_columns_df that do not have a match in the Users_columns_df, those rows will be excluded from the resulting new_df.
and for the missing values I can only either fill them or drop them, 
but I will not be able to get the original values back, so I will have to make a decision on how to handle the missing values based on the context of the data and the analysis I want to perform.
unless NAN means that the post has no views, then I can fill the missing values with 0, but if NAN means that the post has not been viewed yet, then I can fill the missing values with the mean or median of the ViewCount column.
"""
```

```python
new_df['ViewCount'].isnull().mean()
```

```python
posts_columns_df[['ViewCount']]
```

```python
"""
Froim this analysis we can see that the missing values in the ViewCount means that the post was not viewed yet we can fill it with 0.
"""
```

```python
new_df = new_df.fillna(int(0))
new_df.head()
```
